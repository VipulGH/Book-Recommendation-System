## Building a Popularity based Recommender System

```python
ratings_books = ratings.merge(books,on="ISBN")
```

```python
ratings_books
```

```python
freq_rating = ratings_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
freq_rating.rename(columns={'Book-Rating':'rating_count'},inplace=True)
freq_rating
```

```python
freq_avg_rating = ratings_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
freq_avg_rating.rename(columns={'Book-Rating':'rating_avg'},inplace=True)
freq_avg_rating
```

```python
merged_rating_df = freq_rating.merge(freq_avg_rating,on="Book-Title")
merged_rating_df
```

```python
top_50_books = merged_rating_df[merged_rating_df['rating_count']>=250].sort_values('rating_avg',ascending=False).head(50)
```

```python
top_50_books
```

```python
popular_top_50 = top_50_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','rating_avg','rating_count','Image-URL-M']]
```

```python
popular_top_50
```

## Colaborative Filter Based recommender system

```python
temp = ratings_books.groupby('User-ID').count()['Book-Rating']>180
rated_users = temp[temp].index
```

```python
filtered_rating = ratings_books[ratings_books['User-ID'].isin(rated_users)]
```

```python
filtered_rating
```

```python
temp1 = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = temp1[temp1].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
new_data = final_rating.pivot_table(index="Book-Title",columns='User-ID',values="Book-Rating")
```

```python
new_data.fillna(0,inplace= True)
```

```python
new_data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(new_data)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(new_data.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data =[]
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title']==new_data.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
        
    return data 
```

```python
recommend("Harry Potter and the Prisoner of Azkaban (Book 3)")
```
